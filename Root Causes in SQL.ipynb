```python
%sql
CREATE OR REPLACE TEMP VIEW flight_raw
USING csv
OPTIONS (
  path "/Volumes/flight/default/final_single_csv/flight_with_weather_2019/flight_with_weather_2019.csv",
  header "true",
  inferSchema "true"
);
```

```python
%sql
SELECT * FROM flight_raw LIMIT 10;
```

```python
%sql
SELECT
  COUNT(*) AS total_rows,
  COUNT(ARR_DELAY) AS arr_delay_count,
  COUNT(DEP_DELAY) AS dep_delay_count,
  COUNT(TAXI_OUT) AS taxi_out_count,
  COUNT(TAXI_IN) AS taxi_in_count,
  COUNT(AIR_TIME) AS air_time_count
FROM flight_raw;
```

```python
%sql
CREATE OR REPLACE TEMP VIEW flight_clean AS
SELECT
  *,
  CASE WHEN ARR_DELAY < 0 THEN 0 ELSE ARR_DELAY END AS ARR_DELAY_C,
  CASE WHEN DEP_DELAY < 0 THEN 0 ELSE DEP_DELAY END AS DEP_DELAY_C,
  CASE WHEN TAXI_OUT < 0 THEN 0 ELSE TAXI_OUT END AS TAXI_OUT_C,
  CASE WHEN TAXI_IN < 0 THEN 0 ELSE TAXI_IN END AS TAXI_IN_C
FROM flight_raw
WHERE
  ARR_DELAY IS NOT NULL
  AND DEP_DELAY IS NOT NULL
  AND TAXI_OUT IS NOT NULL
  AND TAXI_IN IS NOT NULL
  AND AIR_TIME IS NOT NULL
  AND CRS_ELAPSED_TIME IS NOT NULL;
```

```python
%sql
CREATE OR REPLACE TEMP VIEW flight_kpi AS
SELECT
  *,
  (DEP_DELAY_C + TAXI_OUT_C) AS GROUND_DELAY
FROM flight_clean;
```

```python
%sql
CREATE OR REPLACE TEMP VIEW flight_kpi2 AS
SELECT
  *,
  CASE
    WHEN AIR_TIME - (CRS_ELAPSED_TIME - 25) < 0 THEN 0
    ELSE AIR_TIME - (CRS_ELAPSED_TIME - 25)
  END AS AIR_DELAY
FROM flight_kpi;
```

```python
%sql
CREATE OR REPLACE TEMP VIEW flight_kpi3 AS
SELECT
  *,
  TAXI_IN_C AS ARRIVAL_CONGESTION_DELAY
FROM flight_kpi2;
```

```python
%sql
CREATE OR REPLACE TEMP VIEW flight_kpi_final AS
SELECT
  *,
  CASE
    WHEN O_PRCP > 0 OR O_WSPD > 20 THEN ARR_DELAY_C
    ELSE 0
  END AS WEATHER_DELAY
FROM flight_kpi3;
```

```python
%sql
SELECT
  ARR_DELAY_C,
  GROUND_DELAY,
  AIR_DELAY,
  ARRIVAL_CONGESTION_DELAY,
  WEATHER_DELAY
FROM flight_kpi_final
LIMIT 10;
```

```python
%sql
SELECT
  AVG(ARR_DELAY_C) AS ARRIVAL_KPI,
  AVG(GROUND_DELAY) AS GROUND_KPI,
  AVG(AIR_DELAY) AS AIR_KPI,
  AVG(ARRIVAL_CONGESTION_DELAY) AS ARRIVAL_CONGESTION_KPI,
  AVG(WEATHER_DELAY) AS WEATHER_KPI
FROM flight_kpi_final;
```

```python
%sql
SELECT 'Ground' AS Delay_Type, AVG(GROUND_DELAY) AS Avg_Delay FROM flight_kpi_final
UNION ALL
SELECT 'Air', AVG(AIR_DELAY) FROM flight_kpi_final
UNION ALL
SELECT 'Arrival Congestion', AVG(ARRIVAL_CONGESTION_DELAY) FROM flight_kpi_final
UNION ALL
SELECT 'Weather', AVG(WEATHER_DELAY) FROM flight_kpi_final;
```

```python
%sql
SELECT
  MONTH(FL_DATE) AS MONTH,
  AVG(ARR_DELAY_C) AS AVG_ARR_DELAY
FROM flight_kpi_final
GROUP BY MONTH(FL_DATE)
ORDER BY MONTH;
```

```python
%sql
SELECT
  CASE
    WHEN WEATHER_DELAY > 0 THEN 'Weather'
    ELSE 'No Weather'
  END AS WEATHER_FLAG,
  AVG(ARR_DELAY_C) AS AVG_DELAY
FROM flight_kpi_final
GROUP BY WEATHER_FLAG;
```

```python
%sql
SELECT
  OP_CARRIER,
  AVG(GROUND_DELAY) AS AVG_GROUND_DELAY
FROM flight_kpi_final
GROUP BY OP_CARRIER
ORDER BY AVG_GROUND_DELAY DESC;
```

```python
%sql
CREATE TABLE flight_delay_kpi
USING DELTA
AS
SELECT * FROM flight_kpi_final;
```
