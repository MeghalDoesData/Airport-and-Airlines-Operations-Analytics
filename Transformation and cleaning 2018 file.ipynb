```python
df.createOrReplaceTempView("flights")
```

```python
%sql
-- Business-clean delay values
-- Rule:
-- 1. Early departures/arrivals (< 0) → 0
-- 2. Extreme delays (> 300) → 300
-- 3. Keep values between 0 and 300 as-is

CREATE OR REPLACE TEMP VIEW flights_cleaned AS
SELECT
    *,

    CASE
        WHEN dep_delay < 0 THEN 0
        WHEN dep_delay > 300 THEN 300
        ELSE dep_delay
    END AS dep_delay_clean,

    CASE
        WHEN arr_delay < 0 THEN 0
        WHEN arr_delay > 300 THEN 300
        ELSE arr_delay
    END AS arr_delay_clean

FROM flights;
```

```python
%sql
SELECT
    MIN(dep_delay_clean) AS min_dep,
    MAX(dep_delay_clean) AS max_dep,
    MIN(arr_delay_clean) AS min_arr,
    MAX(arr_delay_clean) AS max_arr
FROM flights_cleaned;
```

```python
%sql
-- Remove records with NULLs in critical analytical columns
-- Keep data that is valid for congestion & efficiency KPIs

CREATE OR REPLACE TEMP VIEW flights_non_null AS
SELECT *
FROM flights_cleaned
WHERE
    origin IS NOT NULL
    AND dest IS NOT NULL
    AND crs_dep_time IS NOT NULL
    AND taxi_out IS NOT NULL
    AND taxi_in IS NOT NULL
    AND dep_delay_clean IS NOT NULL
    AND arr_delay_clean IS NOT NULL;
```

```python
%sql
SELECT COUNT(*) AS total_valid_records
FROM flights_non_null;
```

```python
%sql
-- Remove logically invalid flight records
-- Ensures physical & operational correctness

CREATE OR REPLACE TEMP VIEW flights_valid AS
SELECT *
FROM flights_cleaned
WHERE
    taxi_out >= 0
    AND taxi_in >= 0
    AND air_time > 0
    AND actual_elapsed_time >= air_time
    AND crs_elapsed_time > 0;
```

```python
%sql
SELECT COUNT(*) AS valid_rows
FROM flights_valid;
```

```python
%sql
-- Extract hour from scheduled departure time
-- Enables time-of-day congestion analysis

CREATE OR REPLACE TEMP VIEW flights_time_enriched AS
SELECT
    *,
    HOUR(crs_dep_time) AS dep_hour
FROM flights_valid;
```

```python
%sql
--flights based on time
SELECT
    dep_hour,
    COUNT(*) AS flights
FROM flights_time_enriched
GROUP BY dep_hour
ORDER BY dep_hour;
```

```python
%sql
-- Create a binary delay flag
-- 1 = delayed flight
-- 0 = on-time flight

CREATE OR REPLACE TEMP VIEW flights_with_delay_flag AS
SELECT
    *,
    CASE
        WHEN dep_delay_clean > 0 THEN 1
        ELSE 0
    END AS dep_delay_flag
FROM flights_time_enriched;
```

```python
%sql
SELECT
    dep_delay_flag,
    COUNT(*) AS flights
FROM flights_with_delay_flag
GROUP BY dep_delay_flag;
```

```python
%sql
-- Remove duplicate flight records
-- Keep only one record per logical flight

CREATE OR REPLACE TEMP VIEW flights_deduped AS
SELECT *
FROM (
    SELECT
        *,
        ROW_NUMBER() OVER (
            PARTITION BY
                fl_date,
                op_carrier,
                op_carrier_fl_num,
                origin,
                crs_dep_time
            ORDER BY fl_date
        ) AS rn
    FROM flights_with_delay_flag
)
WHERE rn = 1;
```

```python
%sql
SELECT COUNT(*) AS final_clean_rows
FROM flights_deduped;
```

```python
%sql
-- Airport-level congestion KPIs
-- These metrics quantify how congested each airport is

CREATE OR REPLACE TEMP VIEW airport_kpis AS
SELECT
    origin AS airport,

    COUNT(*) AS total_flights,

    ROUND(AVG(taxi_out), 2) AS avg_taxi_out,
    ROUND(AVG(taxi_in), 2) AS avg_taxi_in,

    -- Delay percentage = delayed flights / total flights
    ROUND(AVG(dep_delay_flag) * 100, 2) AS delay_percentage

FROM flights_deduped
GROUP BY origin;
```

```python
%sql
SELECT *
FROM airport_kpis
ORDER BY delay_percentage DESC
LIMIT 10;
```

```python
%sql
-- ============================================================
-- RECREATE SCORED_AIRPORTS (NORMALIZATION LAYER)
-- ============================================================

CREATE OR REPLACE TEMP VIEW scored_airports AS
WITH stats AS (
    SELECT
        MIN(avg_taxi_out) AS min_taxi_out,
        MAX(avg_taxi_out) AS max_taxi_out,
        MIN(avg_taxi_in)  AS min_taxi_in,
        MAX(avg_taxi_in)  AS max_taxi_in,
        MIN(delay_percentage) AS min_delay,
        MAX(delay_percentage) AS max_delay
    FROM airport_kpis
)
SELECT
    k.airport,
    k.total_flights,
    k.avg_taxi_out,
    k.avg_taxi_in,
    k.delay_percentage,

    -- Normalized metrics (0–1 scale)
    (k.avg_taxi_out - s.min_taxi_out) /
    (s.max_taxi_out - s.min_taxi_out) AS norm_taxi_out,

    (k.avg_taxi_in - s.min_taxi_in) /
    (s.max_taxi_in - s.min_taxi_in) AS norm_taxi_in,

    (k.delay_percentage - s.min_delay) /
    (s.max_delay - s.min_delay) AS norm_delay

FROM airport_kpis k
CROSS JOIN stats s;
```

```python
%sql
-- ============================================================
-- FINAL AIRPORT CONGESTION INTELLIGENCE TABLE
-- (View definition must NOT contain ORDER BY)
-- ============================================================

CREATE OR REPLACE TEMP VIEW airport_congestion_intelligence AS
SELECT
    -- Airport identifier
    s.airport,

    -- Overall congestion metrics
    s.congestion_rank,
    s.congestion_score,
    s.total_flights            AS yearly_flights,
    s.avg_taxi_out,
    s.avg_taxi_in,
    s.delay_percentage         AS overall_delay_percentage,

    -- Peak congestion hour details
    p.dep_hour                 AS peak_hour,
    p.total_flights            AS peak_hour_flights,
    p.delay_percentage         AS peak_hour_delay_percentage

FROM scored_airports s
LEFT JOIN peak_congestion_hours p
    ON s.airport = p.airport;
```

```python
%sql
-- ============================================================
-- PEAK HOUR CONGESTION ANALYSIS
-- This identifies WHEN airports are most congested
-- ============================================================

CREATE OR REPLACE TEMP VIEW airport_hourly_congestion AS
SELECT
    origin AS airport,

    -- Hour of scheduled departure (0–23)
    dep_hour,

    -- Number of flights in this hour
    COUNT(*) AS total_flights,

    -- Percentage of delayed flights in this hour
    ROUND(AVG(dep_delay_flag) * 100, 2) AS delay_percentage

FROM flights_deduped
GROUP BY origin, dep_hour;
```

```python
%sql
-- Show most congested airport-hour combinations
SELECT *
FROM airport_hourly_congestion
ORDER BY delay_percentage DESC
LIMIT 20;
```

```python
%sql
-- ============================================================
-- PEAK CONGESTION HOURS FILTER
-- Keeps only meaningful and actionable congestion windows
-- ============================================================

CREATE OR REPLACE TEMP VIEW peak_congestion_hours AS
SELECT
    airport,
    dep_hour,
    total_flights,
    delay_percentage
FROM airport_hourly_congestion
WHERE
    -- High congestion condition
    delay_percentage >= 60

    -- Avoid small-sample noise
    AND total_flights >= 50
ORDER BY
    delay_percentage DESC;
```

```python
%sql
-- View top peak congestion hours across all airports
SELECT *
FROM peak_congestion_hours
LIMIT 20;
```

```python
%sql
-- ============================================================
-- FINAL AIRPORT CONGESTION SCORE TABLE
-- ============================================================

CREATE OR REPLACE TEMP VIEW airport_congestion_score AS
SELECT
    airport,
    total_flights,
    avg_taxi_out,
    avg_taxi_in,
    delay_percentage,

    -- Final weighted congestion score
    ROUND(
        (norm_taxi_out * 0.4) +
        (norm_taxi_in * 0.3) +
        (norm_delay   * 0.3),
        3
    ) AS congestion_score,

    -- Rank airports by congestion
    RANK() OVER (
        ORDER BY
            (norm_taxi_out * 0.4) +
            (norm_taxi_in * 0.3) +
            (norm_delay   * 0.3) DESC
    ) AS congestion_rank

FROM scored_airports;
```

```python
%sql
-- ============================================================
-- FINAL AIRPORT CONGESTION INTELLIGENCE TABLE
-- (View definition must NOT contain ORDER BY)
-- ============================================================

CREATE OR REPLACE TEMP VIEW airport_congestion_intelligence AS
SELECT
    -- Airport identifier
    s.airport,

    -- Overall congestion metrics
    s.congestion_rank,
    s.congestion_score,
    s.total_flights            AS yearly_flights,
    s.avg_taxi_out,
    s.avg_taxi_in,
    s.delay_percentage         AS overall_delay_percentage,

    -- Peak congestion hour details
    p.dep_hour                 AS peak_hour,
    p.total_flights            AS peak_hour_flights,
    p.delay_percentage         AS peak_hour_delay_percentage

FROM airport_congestion_score s
LEFT JOIN peak_congestion_hours p
    ON s.airport = p.airport;
```
